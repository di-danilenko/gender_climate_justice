```python
# define the query ID we re looking at
q = Query.objects.get(pk=11087)
p = q.project

# get the document IDs attached to this query
dids = set(Doc.objects.filter(query=q).values_list('pk', flat=True))
print(len(dids))
```

```python
# get all the documents identified in adaptation policy project (both seen and unseen)
django.db.connection.close()

def mode_na(x):
    try:
        return statistics.mode([y for y in x if y is not np.NaN])
    except:
        return np.NaN
    
columns = [
    'id',
    'content',
    'title',
    'seen',
    'relevant'
]

dfadaptpolicy = pd.read_csv('finalDf-1.csv', usecols = columns)
display(dfadaptpolicy)
print(len(dfadaptpolicy[dfadaptpolicy['seen'] == 1]))
```

```python
# get documents from the updated query results (jan 23)

final_docs = Doc.objects.filter(
    id__in=dids,
    content__iregex='\w',
)

nacsoscolumns = ['id',
    'content',
    'title',
    'PY',
    'authors',
    'docauthinst__AF'
    'docauthinst', 
                 # how do we include values for all authors at this stage
    'journal'
]

# check whether this works if we do it one by one
for document in final_docs:
    print(dict(document.__dict__('DocAuthInst')))

final_df = pd.DataFrame.from_dict(list(final_docs.values(*nacsoscolumns)))

final_df = final_df[nacsoscolumns]
final_df['relevant'] = 0
final_df['seen'] = 0

print(final_df.shape)
final_df.to_csv('11087_results.csv', index=False)
final_df.head()
```

```python
# write all the docs that have been seen before into seen_df

seen_df = pd.DataFrame(columns=columns)

for x in dfadaptpolicy.index:
    if dfadaptpolicy.iloc[x]['seen'] == 1:
        seen_df.loc[len(seen_df)] = dfadaptpolicy.iloc[x]

    
display(seen_df)
```

```python
seen_df.to_csv('seen.csv',index=False)
```

```python
# write all the docs that have not been seen before into unseen_df

unseen_df = pd.DataFrame(columns=nacsoscolumns)
for y in final_df.index:
    print('im on row', y)
    if final_df.iloc[y]['id'] not in seen_df['id']:
        unseen_df.loc[len(unseen_df)] = final_df.iloc[y]
```
